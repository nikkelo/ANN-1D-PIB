This file is a test implementation of a solution to the problem using the **ridge regression** model. 

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
```

The data is loaded into pandas from a file generated by the Fortran numerical solver, which was then processed by another Python script to convert it to dimensionless format and add headers.

```python
df = pd.read_csv('../dataset/training_data_nodim.csv')
```

Load X as the box length and coefficients, and y as the energies (or ground state energy). Perform the splitting for test and training data.

```python
X = df[["Box Length", "Coefficient 0", "Coefficient 1", "Coefficient 2", "Coefficient 3", "Coefficient 4"]]
y = df[["Energy 0", "Energy 1", "Energy 2", "Energy 3", "Energy 4", "Energy 5", "Energy 6", "Energy 7", "Energy 8", "Energy 9"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

Run the model.

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)

print("R2 score:", r2_score(y_test, y_pred))
```
